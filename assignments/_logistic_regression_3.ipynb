```python
class ManualLogistic:

    def __init__(self, learning_rate=0.01, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.w = None
        self.b = None

    def sigmoid_function(self, value):
        return 1 / (1 + np.exp(-value))

    def train(self, X, y):
        samples, features = X.shape
        self.w = np.zeros(features)
        self.b = 0

        for i in range(self.iterations):

            z = np.dot(X, self.w) + self.b
            predictions = self.sigmoid_function(z)

            gradient_w = (1 / samples) * np.dot(X.T, (predictions - y))
            gradient_b = (1 / samples) * np.sum(predictions - y)

            self.w = self.w - self.learning_rate * gradient_w
            self.b = self.b - self.learning_rate * gradient_b

    def probability(self, X):
        z = np.dot(X, self.w) + self.b
        return self.sigmoid_function(z)

    def classify(self, X, threshold=0.5):
        p = self.probability(X)
        return (p >= threshold).astype(int)

```

```python
X = np.array([[1],[2],[3],[4]])
y = np.array([0,0,1,1])

classifier = ManualLogistic()

classifier.train(X,y)

print(classifier.classify(X))

```
