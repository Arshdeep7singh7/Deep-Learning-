```python
import numpy as np

np.random.seed(0)

inputs = np.linspace(0,10,50).reshape(-1,1)
noise_term = np.random.normal(0,0.2,size=(50,1))

targets = np.log(inputs+1) + noise_term

W_hidden = np.random.uniform(-1,1,size=(1,3))
b_hidden = np.zeros((1,3))

W_output = np.random.uniform(-1,1,size=(3,1))
b_output = np.zeros((1,1))
```

```python
def relu(x):
    return np.maximum(0,x)

def relu_grad(x):
    return (x>0).astype(float)
```

```python
lr = 0.01

for step in range(1000):

    hidden_linear = inputs @ W_hidden + b_hidden
    hidden_activation = relu(hidden_linear)

    predictions = hidden_activation @ W_output + b_output

    diff = predictions - targets
    mse_loss = np.mean(diff**2)

    grad_pred = 2 * diff / len(inputs)

    grad_W_output = hidden_activation.T @ grad_pred
    grad_b_output = np.sum(grad_pred,axis=0,keepdims=True)

    grad_hidden = grad_pred @ W_output.T

    grad_hidden_linear = grad_hidden * relu_grad(hidden_linear)

    grad_W_hidden = inputs.T @ grad_hidden_linear
    grad_b_hidden = np.sum(grad_hidden_linear,axis=0,keepdims=True)

    W_hidden = W_hidden - lr * grad_W_hidden
    b_hidden = b_hidden - lr * grad_b_hidden
    W_output = W_output - lr * grad_W_output
    b_output = b_output - lr * grad_b_output

    if step % 100 == 0:
        print(step,mse_loss)
```
