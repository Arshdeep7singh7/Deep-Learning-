```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import Perceptron
```

```python
# ----- OR Gate Dataset -----
or_table = pd.DataFrame({
    "x1":[0,0,1,1],
    "x2":[0,1,0,1],
    "label":[0,1,1,1]
})

features_or = or_table[["x1","x2"]]
target_or = or_table["label"]

model_or = Perceptron(max_iter=1000,eta0=0.1,random_state=42)
model_or.fit(features_or,target_or)

```

```python
sns.scatterplot(
    data=or_table,
    x="x1",
    y="x2",
    hue="label",
    s=200
)

w_or1, w_or2 = model_or.coef_[0]
bias_or = model_or.intercept_[0]

x_vals = np.linspace(-0.2,1.2,100)
boundary_or = -(w_or1*x_vals + bias_or)/w_or2

plt.plot(x_vals,boundary_or,color="black")
plt.title("OR Gate Decision Boundary")
plt.show()

```

```python
sns.scatterplot(
    data=and_table,
    x="x1",
    y="x2",
    hue="label",
    s=200
)

w_and1, w_and2 = model_and.coef_[0]
bias_and = model_and.intercept_[0]

x_vals = np.linspace(-0.2,1.2,100)
boundary_and = -(w_and1*x_vals + bias_and)/w_and2

plt.plot(x_vals,boundary_and,color="black")
plt.title("AND Gate Decision Boundary")
plt.show()

```
