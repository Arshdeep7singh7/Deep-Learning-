```python
# keep only required columns
features = features.drop(columns=['Sex','Diameter','Whole_weight','Shucked_weight','Viscera_weight'])
features.head()
```

```python
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y = train_test_split(
    features,
    targets,
    test_size=0.2,
    random_state=42
)
```

```python
train_x.shape
test_x.shape
train_y.shape
test_y.shape
```

```python
avg = np.mean(train_x,axis=0)
sd = np.std(train_x,axis=0)
sd[sd==0] = 1

train_x = (train_x-avg)/sd
test_x = (test_x-avg)/sd
```

```python
def linear_forward(data,weights,bias):

    prediction = data @ weights + bias

    print("Shape of X:",data.shape)
    print("Shape of w:",weights.shape)
    print("Shape of b:",np.shape(bias))
    print("Shape of y_hat:",prediction.shape)

    return prediction
```

```python
def compute_mse(y_true,y_pred):

    error = np.mean((y_true-y_pred)**2)
    return error
```

```python
def weight_gradient(data,y_true,y_pred):

    m = data.shape[0]
    grad = (2/m)*(data.T @ (y_pred-y_true))
    return grad


def bias_gradient(y_true,y_pred):

    m = y_true.shape[0]
    grad = (2/m)*np.sum(y_pred-y_true)
    return grad
```

```python
feature_count = train_x.shape[1]

weights = np.random.randn(feature_count)*0.01
bias = 0

lr = 0.01
epochs = 1000
interval = 100

for e in range(epochs):

    pred = linear_forward(train_x,weights,bias)

    loss = compute_mse(train_y,pred)

    gw = weight_gradient(train_x,train_y,pred)
    gb = bias_gradient(train_y,pred)

    weights = weights - lr*gw
    bias = bias - lr*gb

    if e % interval == 0:
        print(f"Epoch {e}, Loss: {loss}")
```

```python
test_pred = linear_forward(test_x,weights,bias)

mse_val = compute_mse(test_y,test_pred)
mae_val = np.mean(np.abs(test_y-test_pred))

print("Test MSE:",mse_val)
print("Test MAE:",mae_val)
```
