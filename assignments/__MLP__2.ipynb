```python
from tensorflow import kerasimport tensorflow as tffrom tensorflow.keras import layersimport numpy as np
```

```python
data_points = np.random.rand(1000,3)target_values = 3 * data_points[:,0] + 2 * data_points[:,1] - data_points[:,2]target_values = target_values + np.random.randn(1000) * 0.1
```

```python
network = keras.Sequential()network.add(layers.Dense(64, activation='relu', input_shape=(3,)))network.add(layers.Dense(32, activation='relu'))network.add(layers.Dense(1))
```

```python
network.compile(    optimizer='adam',    loss='mse',    metrics=['mae'])
```

```python
network.fit(    data_points,    target_values,    epochs=50,    batch_size=32,    validation_split=0.2)
```

```python
mse_score, mae_score = network.evaluate(data_points, target_values)print("MSE:", mse_score)print("MAE:", mae_score)sample_input = np.array([[0.5,0.2,0.1]])predicted_value = network.predict(sample_input)print("Prediction:", predicted_value)
```
